```python
x = np.array([1,2,3,4,5,1,2,1,1,1])
```

```python
count = np.bincount(x)
freq = np.argmax(count)
ind = np.where(x == freq)[0]
print("Most frequent number is: ",freq)
print("Indices of number is: ",ind)
```

```python
y = np.array([1, 1, 1, 2, 3, 4, 2, 4, 3, 3])
```

```python
count = np.bincount(y)
max = np.max(count)
freq = np.where(count == max)[0]
ind = {val: np.where(y == val)[0] for val in freq}
```

```python
print("Frequency: ", freq)
print("Indices: ", ind)
```

```python
from PIL import Image
import numpy as np

def img_to_array(path, save_path):
    img = Image.open(path)

    img_array = np.array(img)
    np.savetxt(save_path, img_array.reshape(-1, img_array.shape[2] if img_array.ndim == 3 else 1), fmt='%d', delimiter=',')

    print(f'Image data saved to {save_path}')

img_to_array('/content/dog.jpeg', 'saved_image_data.txt')
```

```python
import numpy as np
from PIL import Image
import os

def load_array(file_path):
    try:
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"{file_path} does not exist")

        img_array = np.loadtxt(file_path, dtype=int, delimiter=',')
        print(f"Array size: {img_array.size}")
        print(f"Array shape: {img_array.shape}")

        num_pixels = img_array.shape[0]
        possible_dims = []
        for height in range(1, int(np.sqrt(num_pixels)) + 1):
            if num_pixels % height == 0:
                width = num_pixels // height
                possible_dims.append((height, width))
        for height, width in possible_dims:
            try:
                if num_pixels == height * width * 3:
                    img_array = img_array.reshape((height, width, 3))
                    mode = 'RGB'
                    img = Image.fromarray(np.uint8(img_array), mode)
                    return img
            except Exception as e:
                print(f"Failed to reshape to {height}x{width}x3: {e}")

        raise ValueError("Could not determine suitable dimensions for reshaping")

    except Exception as e:
        print(f"An error occurred: {e}")
        return None

img = load_array('saved_image_data.txt')
if img:
    img.show()
else:
    print("No image to show")
```
